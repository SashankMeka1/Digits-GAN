```python
import os
import numpy as np
import gzip
from sklearn.preprocessing import OneHotEncoder
imgs = [os.path.join("Data", file) for file in os.listdir("Data")]
labels = [os.path.join("Labels", file) for file in os.listdir("Labels")]
data = []
print([[x for x in zip(imgs, labels)]])
for images, targets in zip(imgs, labels):
    imgs = gzip.open(images, 'r')
    imgs.read(16)
    buf = imgs.read()
    imgs = np.frombuffer(buf, dtype = np.uint8).astype(np.float32).reshape(-1, 28, 28)/255
    lbls = gzip.open(targets, 'r')
    lbls.read(8)
    buf = lbls.read()
    lbls = np.frombuffer(buf, dtype = np.uint8).astype(np.float32).reshape(-1, 1)
    data.append((imgs, lbls))
```

```python
from matplotlib import pyplot as plt
plt.imshow(data[0][0][0])
print(data[0][1][0])
```

```python
train, test = data[0], data[1]
import torch

from torch import from_numpy as cast
from torch.utils.data import TensorDataset, DataLoader
dataset = TensorDataset(cast(test[0]).float(), cast(test[1]).float())
```

```python
import torch.nn as nn
import torch.nn.functional as F
class generator(nn.Module):
    def __init__(self):
        super(generator, self).__init__()
        self.fc1 = nn.Linear(100, 1024)
        self.fc2 = nn.Linear(1024, 1024)
        self.fc3 = nn.Linear(1024, 1024)
        self.fc4 = nn.Linear(1024, 784)
    def forward(self, x):
        x = self.fc1(x)
        x = F.leaky_relu(x, 0.2)
        x = self.fc2(x)
        x = F.leaky_relu(x, 0.2)
        x = self.fc3(x)
        x = F.leaky_relu(x, 0.2)
        x = self.fc4(x)
        x = torch.tanh(x)
        return x
class discriminator(nn.Module):
    def __init__(self):
        super(discriminator, self).__init__()
        self.fc1 = nn.Linear(784, 1024)
        self.fc2 = nn.Linear(1024, 1024)
        self.fc3 = nn.Linear(1024, 1)
    def forward(self, x):
        x = x.view(-1, 784)
        x = self.fc1(x)
        x = F.leaky_relu(x, 0.2)
        x = self.fc2(x)
        x = F.leaky_relu(x, 0.2)
        x = self.fc3(x)
        x = torch.sigmoid(x)
        return x
a = generator()
b = discriminator()
b(a(torch.randn(3, 100))).shape
```

```python
import torch.optim as optim
from matplotlib import pyplot as plt
import random
gen = generator()
dis = discriminator()
loader = DataLoader(dataset, batch_size = 128, num_workers = 0, shuffle = True)
gopt = optim.Adam(gen.parameters(), lr = 0.0002, betas = (0.5, 0.999))
dopt = optim.Adam(dis.parameters(), lr = 0.0002, betas = (0.5, 0.999))
criterion = nn.BCELoss()
epochs = 200
def show_image():
    plt.imshow(gen(torch.randn(100)).reshape(28, 28).detach().numpy())
    plt.show()
for epoch in range(epochs):
    print("Epoch", epoch+1)
    for i, batch in enumerate(loader):
        dopt.zero_grad()
        true_img = batch[0]
        labels = torch.randn(true_img.shape[0], 100)
        true = torch.ones(true_img.shape[0])
        fake = torch.zeros(true_img.shape[0])
        fake_img = gen(labels)
        fake_pred = dis(fake_img.detach()).view(-1)
        #print(fake_pred)
        true_pred = dis(true_img).view(-1)
        d_loss_p = criterion(fake_pred, fake)
        d_loss_n = criterion(true_pred, true)
        d_loss_p.backward()
        d_loss_n.backward()
        d_loss = d_loss_p + d_loss_n
        dopt.step()
        #print("Dis trained")
        
        gopt.zero_grad()
        fake_pred = dis(fake_img).view(-1)
        g_loss =  criterion(fake_pred, true)
        #print("Gen Loss", g_loss.item())
        #print("Dis Loss", d_loss.item())
        g_loss.backward()
        gopt.step()
        #print("Gen trained")
    print("Gen Loss", g_loss.item())
    print("Dis Loss", d_loss.item())
    show_image()
        
        
        
```
